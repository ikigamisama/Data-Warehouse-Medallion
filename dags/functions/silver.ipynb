```python
import os
import boto3
import io

import pandas as pd
from dotenv import load_dotenv

load_dotenv()

s3_client = boto3.client(
    service_name="s3",
    endpoint_url="http://localhost:9000",
    aws_access_key_id=os.getenv("MINIO_ACCESS_KEY"),
    aws_secret_access_key=os.getenv("MINIO_SECRET_KEY"),
    region_name="us-east-1",
)
```

```python
cust_info_response = s3_client.get_object(Bucket="business.analytics", Key="warehouse/bronze/cust_info.csv")

cust_info_df = pd.read_csv(io.StringIO(cust_info_response["Body"].read().decode('utf-8')))
cust_info_df
```
